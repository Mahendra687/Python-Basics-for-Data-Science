### Operators
* Operators are used to perform simple operations such as additions, comparisons etc on variables and values.
* Python supports the following types of operators, we will be looking at some of the commonly used operators.



#### Arithmetic 


```python
#assing a value to x and y
x = 2 
y = 3
```

```python
# additin and mutiplication
a = x+y
b = x*y
print(a)
print(b)
```

```python
a = 4**2 #exponentiation
print(a)
```

```python
b = 18%7 #modulo
print(b)
```

```python
c = x-y-2 #subtraction
print(c)
```

```python
d = x/y # Division
print(d)
```

```python
e = x//y #floor division
print(e)
```

#### Comparion Operator
* These are used to compare two values
* Gives a boolean result: [true/false]



```python
#numeric calculation
a = 2 < 3
print(a)
```

```python
b = 2 == 3
print(b)
```

```python
c = 2 <= 3
print(c)
```

```python
#String comparisons
e = "rahul" < "rohan"
print(e)
```

#### Logical Operators
* Logical operators are used to combine conditional statements
* Gives a boolean result: [True/False]



```python
x = 6
a = x < 5  and x < 10
print(a)
```

```python
b = x < 10 or x < 4
print(b)
```

```python
c = not(x < 5 and x < 10)
print(c)
```
